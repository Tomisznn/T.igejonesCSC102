```python
import pandas as pd
print('Welcome User to the Admissions Department Application')
#creating an empty list
info = []
number = int(input('How many candidates intend to use this program? '))



      

def computerscience():
    name = input("Enter your name: ")
    score = int(input("Enter your JAMB score: "))
    print("Please answer appropriately with 'yes' or 'no' ")
    waec = input("Did you have at least 5 credits in WASSCE/NECO/GCE Examination(s) including Mathematics and English Language?)")
    print("Please answer appropriately with 'yes' or 'no'")
    interview = input("Did you pass your university admission interview ")
    
    if score >= 230 and waec == 'yes' and interview == 'yes':
        print("Congratulations! You have gained admission!")
        records = {'Name' : [name], 'Course' : ["Computer Science"] , 'Score' : [score], 'Examination Status' : [waec], 'Interview Status' : [interview]}
        df = pd.DataFrame(records)
        df.to_csv('admitted.csv')
    else:
            print("Unfortunately, you have not gained admission into this university.")
            records = {'Name': [name], 'Course' : ["Computer Science"], 'Score' : [score], 'Examination Status' : [waec], 'Interview Status' : [interview]}
            df = pd.DataFrame(records)
            df.to_csv('not-admitted.csv')
    return 

def masscommunication():
    name = input("Enter input your name: ")
    score = int(input("Enter your JAMB score: "))
    print("Please answer appropriately with 'yes' or 'no'")
    waec = input("Did you have at least 5 credits in WASSCE/NECO/GCE Examination(s) including Mathematics and English Language?)")
    print("Please answer appropriately with 'yes' or 'no'")
    interview = input("Did you pass your university admission interview  ")
    
    if score >= 220 and waec == 'yes' and interview == 'yes':
        print("Congratulations! You have gained admission!")
        records = {'Name' : [name], 'Course' : ["Mass Communication"], 'Score' : [score], 'Examination Status' : [waec], 'Interview Status' : [interview]}
        df = pd.DataFrame(records)
        df.to_csv('admitted.csv')
    else:
            print("Unfortunately, you have not gained admission into this university.")
            records = {'Name': [name], 'Course' : ["Mass Communication"], 'Score' : [score], 'Examination Status' : [waec], 'Interview Status' : [interview]}
            df = pd.DataFrame(records)
            df.to_csv('not-admitted.csv')
    return 
for x in range(number): 
    course = input("Enter 1 for the Computer Science Department, and 2 for the Mass Communication Department: ")
    if course == '1':
        records = computerscience()
    elif course == '2':
        records = masscommunication()    
    info.append(records)
```
